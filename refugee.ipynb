```python
import pandas as pd

usa = pd.read_csv('us_refugee.csv')
pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", '{:,}'.format)
usa.head()
```

```python
usa.columns = usa.columns.str.replace(' ','_').str.replace('-','_').str.replace("'", "").str.lower()
usa = usa.rename(columns={'refugees_under_unhcrs_mandate': 'refugee'})
usa = usa.rename(columns={'country_of_origin': 'country'})
usa.info()
```

```python
usa.describe()
```

```python
image = pd.read_csv('image.csv')
image.columns = image.columns.str.replace(' ','_').str.lower()
image = image.rename(columns={'country_name': 'country'})
image.info()
```

```python
usa['refugee_asylum'] = usa.refugee + usa.asylum_seekers
usa.head()
```

```python
pivot = usa.pivot_table(index="country", columns= "year", values="refugee_asylum")
pivot
```

```python
usa_final= pd.merge(pivot, image, how ='left', on="country")
usa_final.head()
```

```python
usa_final.to_csv('usa_output.csv',index=False)
```
